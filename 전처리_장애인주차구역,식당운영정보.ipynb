## 식당운영정보 데이터 전처리

```python
filepath = 'C:/Users/jelly/Desktop/패스트캠퍼스 부트캠프/4. 프로젝트/python_3조/주차장/'
filename = '(주차가능여부컬럼)서울관광재단_식당운영정보_20230111.csv'

raw_data = pd.read_csv(filepath+filename, encoding='cp949')
```

```python
raw_data.head()
```

```python
raw_data.info()
```

```python
res_parking = raw_data[['지역명', '주차가능여부', '식당(ID)']]
```

```python
res_parking = res_parking.replace({'주차가능여부': {'Y':1, 'N':0}})
```

```python
res_parking
```

```python
#주차가능하다고 한 점포수와 전체 점포수 계산
seoul_gu_res_parking = res_parking.groupby('지역명')['주차가능여부'].aggregate(['sum', 'count']).reset_index().rename({'sum':'주차가능점포수', 'count':'전체점포수'}, axis=1)
```

```python
seoul_gu_res_parking
```

```python
seoul_gu_res_parking['주차가능점포비율'] = seoul_gu_res_parking['주차가능점포수']/seoul_gu_res_parking['전체점포수']*100
```

```python
seoul_gu_res_parking
```

```python
seoul_gu_res_parking.to_csv('C:/Users/jelly/Desktop/패스트캠퍼스 부트캠프/4. 프로젝트/python_3조/주차장/서울시 구별 주차가능점포수_전처리.csv',
                encoding='cp949')
```
