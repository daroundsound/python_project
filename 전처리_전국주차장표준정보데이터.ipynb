```python
def combine_2col(col1, col2):
    result = str(col1)
    if not pd.isna(col2):
        result += " " + str(col2)
    return result
```

```python
seoul_datamin = seoul_data.drop(['주차장관리번호','소재지도로명주소', '소재지지번주소', '전체주소', '제공기관명', 
                                 '부제시행구분', '운영요일', '평일운영시작시각', '평일운영종료시각', '토요일운영시작시각',
                                 '토요일운영종료시각', '공휴일운영시작시각', '공휴일운영종료시각','1일주차권요금적용시간', 
                                 '1일주차권요금','결제방법', '특기사항', '관리기관명', '전화번호', '위도', '경도', 
                                '장애인전용주차구역보유여부', '데이터기준일자', '제공기관코드'], axis=1).reset_index(drop=True)
```

```python
seoul_datamin = seoul_datamin.astype({'추가단위요금':float, '월정기권요금':float})
```

```python
pd.pivot_table(seoul_datamin, index='구명', columns='급지구분', values='주차장명', aggfunc='count', margins=True, fill_value=0).reset_index()
```

```python
seoul_datamin['급지'] = seoul_datamin['급지구분'].replace({'1':'주차장_상', '2':'주차장_상', '3':'주차장_중', '4':'주차장_하', '5':'주차장_하', '기타':'주차장_기타'})
```

```python
#주차장 구별 등급별 수 데이터
parking_grade = pd.pivot_table(seoul_datamin, index='구명', columns='급지', values='주차장명', aggfunc='count', fill_value=0).reset_index()
```

```python
#주차장 구별 요금정보별 수 데이터
parking_fee = pd.pivot_table(seoul_datamin, index='구명', columns='요금정보', values='주차장명', aggfunc='count', fill_value=0).reset_index()
```

```python
parking_fee = parking_fee.rename({'무료': '무료주차장', '유료': '유료주차장', '혼합': '혼합주차장'}, axis=1)
```

```python
#주차장 구별 공영/민영 수
parking_public = pd.pivot_table(seoul_datamin, index='구명', columns='주차장구분', values='주차장명', aggfunc='count', fill_value=0).reset_index()
```

```python
#주차장 구별 주차구획수
parking_amount = seoul_datamin.groupby('구명')['주차구획수'].sum().reset_index().rename({'주차구획수':'총주차구획수'}, axis=1)
```

```python
parking_public = parking_public.rename({'공영': '공영주차장', '민영': '민영주차장'}, axis=1)
```
