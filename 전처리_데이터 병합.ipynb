## 주차 수요 및 공급 관련 데이터 통합

```python
filepath = 'C:/Users/jelly/Desktop/패스트캠퍼스 부트캠프/4. 프로젝트/python_3조/주차장/전처리 완료/'

filename = '서울시 구별 주차가능점포수_전처리.csv'
res_parking = pd.read_csv(filepath+filename, encoding='cp949')

filename = '장애인주차구역과태료_전처리.csv'
fine = pd.read_csv(filepath+filename, encoding='cp949')

filename = '서울_점포수.csv'
place = pd.read_csv(filepath+filename, encoding='cp949')
```

```python
res_parking = res_parking.drop('Unnamed: 0', axis=1)
```

```python
fine = fine.drop('Unnamed: 0', axis=1)
```

```python
fine.columns
```

```python
place = place.drop('Unnamed: 0', axis=1)
```

```python
place = pd.merge(res_parking, place, left_on='지역명', right_on='행정구역').drop('행정구역', axis=1).rename({'주차가능점포수':'주차가능식당수','전체점포수':'전체식당수', '주차가능점포비율':'주차가능식당(%)','점포수 2023':'전체점포수' }, axis=1)
```

```python
supply = pd.merge(fine, place,  left_index=True, right_index=True).drop('지역명', axis=1)
```

## merge 활용 순차적으로 데이터 붙이기
- 합칠 데이터 목록
  - parking_all 
  - pop 
  - traffic_final 
  - supply
  - trans
  - traffic4
  - law

```python
#순차적으로 데이터 붙이기
data_set = pd.merge(parkinglot, pop, left_on='구명', right_on='행정구역').drop('행정구역', axis=1)
data_set = pd.merge(data_set, traffic_final, left_on='구명', right_on='시군구명').drop('시군구명', axis=1)
data_set = pd.merge(data_set, supply, left_index=True, right_index=True).drop('시군구명', axis=1)
data_set = pd.merge(data_set, trans, left_index=True, right_index=True).drop('구별구분', axis=1)
data_set = pd.merge(data_set, traffic4, left_index=True, right_index=True).drop('구별구분', axis=1)
data_set = pd.merge(data_set, law, left_index=True, right_index=True).drop('구별구분', axis=1)
```
