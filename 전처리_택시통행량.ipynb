## 주요 상권별 택시통행량 전처리

```python
filepath = 'C:/Users/jelly/Desktop/패스트캠퍼스 부트캠프/4. 프로젝트/python_3조/주차장/'
filename = '주요 상권별 택시통행량 정보_2024-02-13.csv'

raw_data = pd.read_csv(filepath+filename, encoding='cp949')
```

```python
raw_data.info()
```

```python
#원하는 데이터 확인
raw_data['signgu_nm'].unique()
```

```python
#시군구 코드를 이용하여 서울 지역 데이터만 추출
seoul_data = raw_data[raw_data['signgu_cd']<20000]
```

```python
seoul_data
```

```python
#25개 구 데이터 확인
seoul_data['signgu_nm'].unique()
```

```python
seoul_data = seoul_data.drop(['trdar_no', 'trdar_nm', 'signgu_cd', 'trdar_dgrm_crdnt'], axis=1).reset_index(drop=True)
```

```python
seoul_data
```

- link_cnt 링크수: 표준노드링크 기준 인근 도로수
- minm_txtr 최소 택시통행량: 각 링크별 평균택시통행량 중 최솟값
- max_txtr 최대 택시통행량: 각 링크별 평균택시통행량 중 최댓값
- ave_txtr 평균 택시통행량: 각 링크별 평균택시통행량의 평균값 

```python
#시군구명, 평균 택시 통행량만 남기기로 함
seoul_taxi = seoul_data.groupby('signgu_nm')['ave_txtr'].mean().reset_index().rename({'signgu_nm':'시군구명'}, axis=1)
```
