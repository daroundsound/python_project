```python
# 미개발 토지 파일 경로 지정
file_path_est = '/content/drive/MyDrive/Python Project_3조/주차장 관련 rawdata/대도시 미개발 토지 정보 [2023-12-22].csv'

# CSV 파일 불러오기
df_est = pd.read_csv(file_path_est)

# 데이터 확인
print(df_est.head())
```

```python
# 한글 칼럼명으로 변경할 딕셔너리 생성
한글_칼럼명 = {
    'pnu': '필지고유번호',
    'signgu_cd': '시군구코드',
    'emdl_cd': '읍면동리코드',
    'cll': '지번구분',
    'hmno': '본번',
    'vcno': '부번',
    'tllbadd': '지번주소',
    'lnct_cd': '지목코드',
    'lnct_nm': '지목명',
    'ldar': '토지면적',
    'posesn_tpcl': '소유구분코드',
    'posesn_clsf_nm': '소유구분명',
    'owner_cnt': '소유인수',
    'susar_cd_one': '용도지역코드1',
    'susar_nm_one': '용도지역명1',
    'susar_cd_two': '용도지역코드2',
    'susar_nm_two': '용도지역명2',
    'land_use_sttn_cd': '토지이용상황코드',
    'land_use_sttn_nm': '토지이용상황명',
    'land_lvl_clsf_cd': '토지고저구분코드',
    'land_lvl_clsf_nm': '토지고저구분명',
    'tpgrp_clsf_cd': '지형구분코드',
    'tpgrp_clsf_nm': '지형구분명',
    'rdsd_cd': '도로접면코드',
    'rdsd_nm': '도로접면명',
    'strd_yr': '기준연도',
    'stdmt': '기준월',
    'olnlp': '공시지가'
}

# 열 이름 변경 적용
df_est.rename(columns=한글_칼럼명, inplace=True)

# 결과 확인
print(df_est.head())
```

```python
# '서울시'로 시작하는 지번주소가 있는지 확인
seoul_addresses = df_est[df_est['지번주소'].str.startswith('서울시')]['지번주소']

# 결과 출력
print(seoul_addresses)
```

```python
# '서울특별시'로 시작하는 행만 남기기
df_est = df_est[df_est['지번주소'].str.startswith('서울특별시')]

# 결과 확인
print(df_est['지번주소'])
```

```python
# '지번주소' 열에서 각 행의 값을 가져옵니다.
addresses = df_est['지번주소']

# 각 주소에서 앞에 세 어절을 추출하여 '시', '구', '상세주소'로 나눕니다.
시_list = []
구_list = []
상세주소_list = []

for address in addresses:
    address_split = address.split()  # 공백을 기준으로 분리
    시 = address_split[0]  # 첫 번째 어절은 '시'
    구 = address_split[1]  # 두 번째 어절은 '구'
    상세주소 = ' '.join(address_split[2:])  # 세 번째 어절부터는 '상세주소'로 합침
    시_list.append(시)
    구_list.append(구)
    상세주소_list.append(상세주소)

# 새로운 열에 값을 추가합니다.
df_est['시'] = 시_list
df_est['구'] = 구_list
df_est['상세주소'] = 상세주소_list

# 결과 확인
print(df_est)
```

```python
# 필요 없는 열을 제거합니다.
df_est.drop(columns=[ '필지고유번호', '시군구코드', '읍면동리코드', '지번구분', '본번', '부번','지목코드','posesn_tpcd','소유인수','용도지역코드1','용도지역코드2','용도지역명2','토지이용상황코드','land_lvlg_clsf_cd'], inplace=True)

# 결과 확인
print(df_est.head())
```

```python
# '지번주소', '지형구분코드', '도로접면코드', '기준연도', '기준월' 열을 삭제
df_est.drop(columns=['지번주소', '지형구분코드', '도로접면코드', '기준연도', '기준월'], inplace=True)

print(df_est.head())
```

```python
# '시', '구', '상세주소' 열을 맨 앞으로 이동
cols_to_move = ['시', '구', '상세주소']
df_est = df_est[cols_to_move + [col for col in df_est.columns if col not in cols_to_move]]

print(df_est.head())
```

```python
#이름바꾸기
df_est.rename(columns={'land_lvlg_clsf_nm': '토지고저구분명'}, inplace=True)

print(df_est.head())
```

```python
#이름바꾸기
df_est.rename(columns={'용도지역명1': '용도지역명'}, inplace=True)

print(df_est.head())
```

```python
#토지매입비 추가
#토지매입비는 공시지가의 1.2배 기준에 주택외 취득세 4.6%적용
df_est['토지매입비'] = df_est['토지면적'] * df_est['공시지가']*1.2*1.046

print(df_est.head())
```

```python
#주차장의 용적율은 1500퍼센트 이하, 대지면적의 최소한도 45제곱미터,
#높이제한은 도로접면에 따라 구분이지만 도로 면적 데이터가 없으므로 고려X
#하지만 세로(불), 세각(불), 맹지 는 자동차 출입이 불가능하므로 제거
df_est = df_est[~df_est['도로접면명'].str.endswith('(불)')] #세로(불),세각(불),맹지 제거
df_est = df_est[df_est['도로접면명'] != '맹지']
df_est = df_est[df_est['토지면적'] >= 45] #대지면적이 45미만인 항목 제거

print(df_est)
```

```python
#지상연면적 열 추가(용적률 1500%)
df_est['지상 연면적'] = df_est['토지면적'] * 15

print(df_est)
```

```python
#주차전용시설은 건축물 면적의 95%이상을 주차시설로 사용해야함
df_est['지상 주차시설 최소면적'] = df_est['지상 연면적'] * 0.95

print(df_est)
```

```python
#공사비 산정은 층별로 공사비 산정을 다르게 해야하지만 편의상 모든층 평당 2,000,000원으로 산정
df_est['예상 공사비'] = df_est['지상 연면적'] * (2000000 / 3.3)

print(df_est)
```

```python
#법정주차구획 면적은 11.5이고
#주차시설의 차량 이동면적 까지 고려해서 1대당 소요 면적을 대략적으로 23으로 선정
df_est['예상 주차가능대수'] = round(df_est['지상 주차시설 최소면적'] / 23, 0)

print(df_est)
```

```python
#용도지역명 종류 확인
unique_values = df_est['용도지역명'].unique()
unique_values

#주차시설의 경우 녹지지역에는 설치 불가능 이므로 자연녹지지역 보전녹지지역 제거
df_est = df_est[~df_est['용도지역명'].isin(['자연녹지지역', '보전녹지지역'])]

print(df_est.head())
```

```python
# '주차 수/비용' 열을 계산하고 추가합니다.
df_est['비용/주차 수'] = (df_est['토지매입비'] + df_est['예상 공사비'])/df_est['예상 주차가능대수']

# 결과를 확인합니다.
print(df_est.head())
```

```python
df_est_sorted = df_est.sort_values(by='비용/주차 수', ascending=True)

# 정렬된 데이터프레임을 출력합니다.
print(df_est_sorted)
```
