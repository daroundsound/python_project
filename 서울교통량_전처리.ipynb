```python
# 교통량 파일 경로 지정
file_path_cars = '/content/drive/MyDrive/Python Project_3조/주차장 관련 rawdata/2024_02_월별_서울교통량.csv'

# CSV 파일 불러오기
df_cars = pd.read_csv(file_path_cars)

# 데이터 확인
print(df_cars.head())
```

```python
# 교통량 측정 위치 파일 경로 지정
file_path_cars_spot = '/content/drive/MyDrive/Python Project_3조/주차장 관련 rawdata/2024_02_서울교통량_위치.csv'

# CSV 파일 불러오기
df_cars_spot = pd.read_csv(file_path_cars_spot)

# 데이터 확인
print(df_cars_spot.head())
```

```python
# 두 데이터프레임을 지점번호를 기준으로 머지
merged_df = pd.merge(df_cars, df_cars_spot, on='지점번호', how='inner')

# 결과를 확인하기 위해 출력
print(merged_df)
```

```python
# 주소 열에서 세 어절을 추출하여 "주소(1)", "주소(2)", "주소(3)" 열 생성
주소_split = merged_df['주소'].str.split(n=2, expand=True)
merged_df['주소(1)'] = 주소_split[0]
merged_df['주소(2)'] = 주소_split[1]
merged_df['주소(3)'] = 주소_split[2]

# 결과 확인
print(merged_df)
```

```python
# 주소(1), 주소(2), 주소(3) 열 추출
주소1 = merged_df['주소(1)']
주소2 = merged_df['주소(2)']
주소3 = merged_df['주소(3)']

# 주소(1), 주소(2), 주소(3) 열을 데이터프레임의 처음 열로 이동
merged_df.drop(['주소(1)', '주소(2)', '주소(3)'], axis=1, inplace=True)
merged_df.insert(0, '주소(3)', 주소3)
merged_df.insert(0, '주소(2)', 주소2)
merged_df.insert(0, '주소(1)', 주소1)

# 결과 확인
print(merged_df)

```

```python
# 도로명 주소 열 제거
merged_df = merged_df.drop(['도로명 주소'], axis=1)

# "주소" 열 제거
merged_df.drop('주소', axis=1, inplace=True)

# 결과 확인
print(merged_df)
```

```python
# 0시부터 계까지의 열들을 제외한 열들의 리스트 작성
cols_except_time = [col for col in merged_df.columns if col not in ['0시', '1시', '2시', '3시', '4시', '5시', '6시', '7시', '8시', '9시', '10시', '11시', '12시', '13시', '14시', '15시', '16시', '17시', '18시', '19시', '20시', '21시', '22시', '23시', '계']]

# 0시부터 계까지의 열들을 순서대로 추가하여 새로운 열의 순서 리스트 작성
new_cols = cols_except_time + ['0시', '1시', '2시', '3시', '4시', '5시', '6시', '7시', '8시', '9시', '10시', '11시', '12시', '13시', '14시', '15시', '16시', '17시', '18시', '19시', '20시', '21시', '22시', '23시', '계']

# 새로운 열의 순서로 데이터프레임 재구성
merged_df = merged_df[new_cols]

# 결과 확인
print(merged_df)
```

```python
# '입출 방향' 열 생성
merged_df['입출 방향'] = merged_df.apply(lambda row: row['유입 방향'] if row['방향'] == '유입' else row['유출방향'], axis=1)

# '유출 방향'과 '유입 방향' 열 삭제
merged_df = merged_df.drop(['유출방향', '유입 방향'], axis=1)

# 결과 확인
print(merged_df)
```

```python
# '지점명칭' 열 추출
지점명칭_column = merged_df['지점명칭']

# '지점명칭' 열 삭제
merged_df.drop('지점명칭', axis=1, inplace=True)

# '지점명칭' 열을 데이터프레임의 처음 열로 삽입
merged_df.insert(0, '지점명칭', 지점명칭_column)

# 결과 확인
print(merged_df)
```

```python
# '구분' 열 제거
merged_df.drop('구분', axis=1, inplace=True)

# 결과 확인
print(merged_df)
```

```python
# '입출방향' 열 추출
입출방향_column = merged_df['입출 방향']

# '입출방향' 열 삭제
merged_df.drop('입출 방향', axis=1, inplace=True)

# '입출방향' 열을 데이터프레임의 8번째 열로 삽입
merged_df.insert(8, '입출 방향', 입출방향_column)

# 결과 확인
print(merged_df)
```

```python
# '지점명' 열 제거
merged_df.drop('지점명', axis=1, inplace=True)

# 결과 확인
print(merged_df)
```

```python
# '검지기 유형' 열 제거
merged_df.drop('검지기 유형', axis=1, inplace=True)

# 결과 확인
print(merged_df)
```

```python
# 시로 끝나는 열을 찾습니다.
columns_to_drop = [col for col in merged_df.columns if col.endswith('시')]

# 해당 열을 삭제한 새로운 데이터프레임을 생성합니다.
new_df = merged_df.drop(columns=columns_to_drop)

# 결과를 출력합니다.
new_df
```
