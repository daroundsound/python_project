```python
import numpy as np
import pandas as pd
```

```python
filepath = 'C:/Users/jelly/Desktop/패스트캠퍼스 부트캠프/4. 프로젝트/python_3조/주차장/'
filename = '서울시 공영주차장 안내 정보.csv'
public = pd.read_csv(filepath+filename, encoding='cp949')
```

```python
# 데이터 컬럼 확인
public.columns
```

```python
# 데이터 값 확인
public.head()
```

```python
# 필요한 데이터만 가져오기
public = public[['주차장코드', '주차장명','총 주자면','주소', '위도', '경도']]
```

```python
public
```

```python
# 주소 데이터 분리해서 자치구명 추출
public['구명'] = public['주소'].str.split(" ", expand=True).iloc[:,[0]]
```

```python
#구획당 한 행으로 구분되어 있는 데이터 확인 -> 주차장코드로 groupby해서 합계 구한 뒤에 진행
#(-> 공영주차장 행 수가 18000여개였던 이유)
public.sort_values(by='총 주자면')
```

```python
#자치구별 총 주차면수 계산
public_pre = public.groupby(['주차장코드', '주차장명', '구명'])['총 주자면'].sum().reset_index()
```

```python
public_pre
```

```python
# 합계 계산한 데이터셋과 원 데이터를 주차장 코드로 붙여서 합계 데이터에 해당 주차장의 주소지 붙이기
public_locate = pd.merge(public_pre, public, on='주차장코드', how='inner').drop(['주차장명_y', '총 주자면_y', '구명_y'], axis=1).rename({'주차장명_x':'주차장명', '구명_x':'구명','총 주자면_x':'총 주차면'}, axis=1)
```

```python
#테스트 데이터 삭제
public_locate = public_locate.drop_duplicates(['주차장코드']).drop([18764]).reset_index(drop=True)
```
